```python
from astropy.table import Table, vstack, Column
from astropy.io import fits
from astropy.io import ascii
import glob
import os
import pandas as pd
```

```python
# Define the data directory
data_dir = '/home/d/dltucker/DATA/S_STDS-exPhotom'
```

```python
# Create list of all exPhotom files
fits_files = glob.glob(os.path.join(data_dir, 'exPhotom-*.fit'))
```

```python
# Initialize an empty list to store tables
tables = []

# Read each FITS file and append its table to the list
for fname in fits_files:

    # Extract MJD from filename
    mjd = float(os.path.basename(fname).split('-')[1].split('.')[0])
    
    with fits.open(fname) as hdul:
        
        table = Table.read(hdul[2])  # Read HDU[2]
 
        # Create new table with MJD as first column
        new_table = Table()
        new_table['MJD'] = Column([mjd] * len(table))
        
        # Add all other columns from original table
        for col in table.colnames:
            new_table[col] = table[col]
        tables.append(new_table)

# Combine all tables vertically into one large table
combined_table = vstack(tables)
```

```python
combined_table
```

```python
# Write combined_table to data_dir as an ECSV file
output_fname = os.path.join(data_dir, 'exPhotom-S_STDS_ALL.ecsv')
combined_table.write(output_fname, overwrite=True)
```
